# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct"]
values = [42, 44, 43, 45, 47, 46, 49, 55, 54, 56]

x = np.arange(len(months))

fig, ax = plt.subplots(figsize=(11, 5))

# Line
ax.plot(x, values, linewidth=3)

# Highlight one important point
highlight_idx = 7
ax.scatter(x[highlight_idx], values[highlight_idx], s=100, zorder=3)

# Minimal styling
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["bottom"].set_visible(False)

ax.tick_params(axis='x', length=0, labelsize=11)
ax.tick_params(axis='y', length=0, labelsize=10)

# Light horizontal gridlines
ax.yaxis.grid(True, linestyle='-', alpha=0.2)
ax.xaxis.grid(False)

# Ticks
ax.set_xticks(x)
ax.set_xticklabels(months)

# Title and subtitle
ax.set_title("Sales rose sharply in late summer", loc="left", fontsize=18, weight="bold")

# Direct annotation
ax.annotate(
    "Notice the jump in August",
    xy=(x[highlight_idx], values[highlight_idx]),
    xytext=(x[highlight_idx] - 2.2, values[highlight_idx] + 5),
    arrowprops=dict(arrowstyle="-", lw=1.5),
    fontsize=11
)

# Optional axis range for cleaner look
ax.set_ylim(38, 60)

plt.tight_layout()
plt.show()
```
